```python
import folium
```

```python
mapa_leon = folium.Map(
                location=[21.0977953, -101.88304764]
            )
```

```python
mapa_leon
```

```python
import pandas as pd

df = pd.read_excel('puntosconectividadleonjun2019.xlsx')
```

```python
df.tail()
```

```python
lat_long = df[['LATITUD', 'LONGITUD']].values
```

```python
etiquetas = df[['PUNTO_DE_CONEXION']].values
```

```python
for i in range(len(etiquetas)):
    folium.Marker(lat_long[i], popup=etiquetas[i][0], tooltip=etiquetas[i][0]).add_to(mapa_leon)
```
